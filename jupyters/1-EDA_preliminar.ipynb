# Análisis preliminar del conjunto de datos

```python
# Importaciónes
import pandas as pd
pd.set_option('display.max_columns', None)
```

# Analizamos archivo bank-additional

```python
# Leer el archivo bank-additional
df_rawBank = pd.read_csv("../data/orig/bank-additional.csv", index_col=0)
df_rawBank.head()
```

```python
df_rawBank.shape
```

```python
df_rawBank.columns
```

```python
df_rawBank.info()
```

```python
#Crear un dataframe con el que trabajar
df_Bank = df_rawBank.copy()
```

```python
#Cambiar el nombre de las columnas
df_Bank.columns = df_Bank.columns.str.replace(" ", "_").str.lower().str.replace(".", "_")
df_Bank.rename(columns={"id_": "id"}, inplace=True)
df_Bank.sample()
```

# Analizamos archivo customer-details

```python
# Leer el archivo customer-details
# Hoja 1
df_rawCustomer1 = pd.read_excel("../data/orig/customer-details.xlsx", sheet_name=0, index_col=0)
df_rawCustomer1.head()
```

```python
df_rawCustomer1.shape
```

```python
# Hoja 2
df_rawCustomer2 = pd.read_excel("../data/orig/customer-details.xlsx", sheet_name=1, index_col=0)
df_rawCustomer2.head()
```

```python
df_rawCustomer2.shape
```

```python
# Hoja 3
df_rawCustomer3 = pd.read_excel("../data/orig/customer-details.xlsx", sheet_name=2, index_col=0)
df_rawCustomer3.head()
```

```python
df_rawCustomer3.shape
```

```python
# Juntamos en el df_rawCustomer12 las hojas 1 y 2
df_rawCustomer12 = pd.concat([df_rawCustomer1, df_rawCustomer2], axis=0, join='outer')
df_rawCustomer12.head()
```

```python
# Agrupamos en el df_rawCustomer el df_rawCustomer12 y la hoja 3
df_rawCustomer = pd.concat([df_rawCustomer12, df_rawCustomer3], axis=0, join='outer')
df_rawCustomer.head()
```

```python
df_rawCustomer.info()
```

```python
df_rawCustomer.columns
```

```python
#Crear un dataframe con el que trabajar
df_customer = df_rawCustomer.copy()
```

```python
#Cambiar el nombre de las columnas
df_customer.columns = df_customer.columns.str.replace(" ", "_").str.lower().str.replace(".", "_")
df_customer.rename(columns={"ID": "id"}, inplace=True)
df_customer.sample()
```

```python
#unimos los dos ficheros
df_concat = pd.merge(df_Bank, df_customer, on="id", how="left")
```

```python
# Guardo el fichero en un una nueva hoja excel
df_concat.to_csv("../data/transformados/bank-customers-detail.csv", index=False)
```
