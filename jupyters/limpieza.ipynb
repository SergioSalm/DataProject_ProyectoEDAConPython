# Transformación de datos

```python
#Eliminamos las columnas 'latitude' y 'longitude'.
df.drop(columns=["latitude"], inplace=True)
df.drop(columns=["longitude"], inplace=True)
```

```python
#pasamos todas las palabras a minúscula
sp.valores_minus(df)
```

```python
df.sample()
```

```python
#Cambiamos los puntos por comas
col_comas = ['cons_price_idx', 'cons_conf_idx', 'euribor3m', 'nr_employed']
sp.comas(df, col_comas)
```

```python
df.sample()
```

```python
#Convertimos el mes de la columna date, que está como string y en español a número entero.
meses = {'enero':'01', 
         'febrero':'02', 
         'marzo':'03', 
         'abril':'04', 
         'mayo':'05', 
         'junio':'06', 
         'julio':'07', 
         'agosto':'08', 
         'septiembre':'09', 
         'octubre':'10', 
         'noviembre':'11', 
         'diciembre':'12', }

df.replace({'date':meses}, regex = True, inplace = True)
```

```python
df.sample()
```

```python
#Convertimos columnas objet a dateTime con el formato de cada una
fecha_formato = {'date':'%d-%m-%Y',
                 'dt_customer':'%Y-%m-%d'}
sp.convertir_fechas(df, fecha_formato)

#Convertimos columnas objet a float o int
sp.convertir_enteros(df)

#Convertimos las columnas con 0,1, nan a 'Si', 'No', 'None'
bin_cols = ["default", "housing", "loan"]
sp.convertir_YesNo(df, bin_cols)

#Reemplazamos el valor 999 por -1 para tener los datos más agrupados
df["pdays"] = df["pdays"].replace(999, -1)

```

```python
df.sample()
```

```python
df.info()
```

# Creamos columnas

```python
#Creamos campos de año y mes para la columna date
df['contact_year'] = df['date'].dt.year
df['contact_month'] = df['date'].dt.month

#Creamos campos de año y mes para la columna dt_customer
df['customer_year'] = df['dt_customer'].dt.year
df['customer_month'] = df['dt_customer'].dt.month

#Creamos una columna nueva para tener un título más aclarativo y borramos la original
df["subscribed"] = df["y"]
df.drop(columns=["y"], inplace=True)

#Creamos la columna duration_min y eliminanos la columna duration
df["duration_min"] = (df["duration"] / 60).round(2)
df.drop(columns=["duration"], inplace=True)
```
