```python
#Convertimos el mes de la columna date, que está como string y en español a número entero.
meses = {'enero':'01', 
         'febrero':'02', 
         'marzo':'03', 
         'abril':'04', 
         'mayo':'05', 
         'junio':'06', 
         'julio':'07', 
         'agosto':'08', 
         'septiembre':'09', 
         'octubre':'10', 
         'noviembre':'11', 
         'diciembre':'12', }

df.replace({'date':meses}, regex = True, inplace = True)
```

```python
#Convertimos columnas
fecha_formato = {'date':'%d-%m-%Y',
                 'dt_customer':'%Y-%m-%d'}
sp.convertir_fechas(df, fecha_formato)

sp.convertir_enteros(df)

bin_cols = ["default", "housing", "loan"]
sp.convertir_YesNo(df, bin_cols)
```

```python
#Creamos campos de año, mes y día para la columna date
df['year_bank'] = df['date'].dt.year
df['month_bank'] = df['date'].dt.month
df['day_bank'] = df['date'].dt.day_name().str.lower()

#Creamos campos de año, mes y día para la columna dt_customer
df['year_customer'] = df['dt_customer'].dt.year
df['month_customer'] = df['dt_customer'].dt.month
df['day_customer'] = df['dt_customer'].dt.day_name().str.lower()
```
